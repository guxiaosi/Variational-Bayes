```python
#Gradient Ascent with Steepest size(Quadratic function)
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    x1= x[0].item()
    x2= x[1].item()
    x3= x[2].item()
    y= -1/2*(3*x1**2+4*x2**2+3*x3**2+4*x2*x3+2*x1*x3)+(1*x1+2*x2+2*x3)
    return y

def grad_f(x):
    x1= x[0].item()
    x2= x[1].item()
    x3= x[2].item()
    g= np.zeros((3,1))
    g[0] = -(3*x1+x3-1)
    g[1] = -(4*x2+2*x3-2)
    g[2] = -(x1+2*x2+3*x3-2)
    return g

Q = np.array([3,0,1,0,4,2,1,2,3]).reshape(3,3)
b = np.array([1,2,2]).reshape(3,1)

eps0 = 1e-3
stepsize = 1e-1
stop=False
itera=0
x0 = np.array([1,1,1]).reshape(3,1)
x = np.zeros((3,1))

while not stop:
    g = grad_f(x0)
    stepsize = (g.T@g)/(g.T@Q@g)#steepest size
    x = x0 + stepsize*g
    if np.linalg.norm(x-x0)/np.linalg.norm(x0)<eps0:
        stop=True
        break
    x0 =x
    itera+=1
print("argmax x: ", x, "iteration: ", itera)
print(np.linalg.inv(Q)@b)
print(f(x))
```
