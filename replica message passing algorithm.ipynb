```python
# Python program showing Graphical
# representation of tanh() function
import numpy as np
import matplotlib.pyplot as plt
  
in_array = np.linspace(-np.pi, np.pi, 19)
out_array = np.tanh(1+2*in_array)
out_array2 = in_array
  
print("in_array : ", in_array)
print("\nout_array : ", out_array)
  
# red for numpy.tanh()
plt.plot(in_array, out_array, color = 'red', marker = "o")
plt.plot(in_array, out_array2, color = 'blue', marker = "o")
plt.title("numpy.tanh()")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
%matplotlib inline
import math
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats


mu_1 = 2
mu_2 = -2
variance = 1
sigma = math.sqrt(variance)
x = 0.5*np.linspace(mu_1 - 3*sigma, mu_1 + 3*sigma, 100)+0.5*np.linspace(mu_2 - 3*sigma, mu_2 + 3*sigma, 100)
plt.plot(x, scipy.stats.norm.pdf(x, mu_1, sigma),scipy.stats.norm.pdf(x, mu_2, sigma))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

mu = [-2, 2]
sigma = [1, 1]
p_i = [0.5, 0.5]
n = 1000

x = []
y = []
for i in range(n):
    z_i = np.argmax(np.random.multinomial(1, p_i))
    x_i = np.random.normal(mu[z_i], sigma[z_i])
    x.append(x_i)

def univariate_normal(x, mean, variance):
    """pdf of the univariate normal distribution."""
    return ((1. / np.sqrt(2 * np.pi * variance)) * 
            np.exp(-(x - mean)**2 / (2 * variance)))

a = np.arange(-6, 6, 0.01)
y = p_i[0] * univariate_normal(a, mean=mu[0], variance=sigma[0]**2) + p_i[1] * univariate_normal(a, mean=mu[1], variance=sigma[0]**2)
#z = y +np.random.normal(0,1)
fig, ax = plt.subplots(figsize=(8, 4))

ax.hist(x, bins=100, density=True)
ax.plot(a, y)
```
